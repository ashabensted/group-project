```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_density_class_over_time(data, title='Density Class Distribution Over Time'):
    """
    Plots a histogram of Density Class over time for a given dataset.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'year' and 'Density Class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the plot.
    """
    # Group data by 'year' and 'Density Class', count occurrences
    density_class_over_time = data.groupby(['year', 'density class']).size().reset_index(name='Count')

    # Pivot the table to create a bar chart-friendly format
    pivot_density_class = density_class_over_time.pivot(index='year', columns='density class', values='Count').fillna(0)

    # Sort density classes (optional, customize as needed)
    density_order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']  # Customize order
    pivot_density_class = pivot_density_class[density_order]

    # Plot
    fig, ax = plt.subplots(figsize=(18, 10))
    pivot_density_class.plot(kind='bar', stacked=True, ax=ax, colormap='tab10', alpha=0.85)

    # Add annotations (optional)
    for container in ax.containers:
        labels = [f'{int(v)}' if v > 0 else '' for v in container.datavalues]
        ax.bar_label(container, labels=labels, label_type='center', fontsize=8)

    # Add titles and labels
    plt.title(title, fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Number of Occurrences', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)  # Rotate and adjust x-axis labels
    plt.legend(title='Density Class', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

    # Adjust gridlines for readability
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()
    plt.close('all')  # Clean up
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_density_class_over_time(data, title='Density Class Distribution Over Time'):
    """
    Plots a line graph of Density Class over time for a given dataset, using intuitive colors for the density classes.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'year' and 'density class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the plot.
    """
    # Define the desired order of density classes
    density_order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

    # Ensure 'density class' is a categorical column with the specified order
    data['density class'] = pd.Categorical(data['density class'], categories=density_order, ordered=True)

    # Group data by 'year' and 'Density Class', count occurrences
    grouped_data = data.groupby(['year', 'density class']).size().reset_index(name='Count')

    # Pivot the table for line plot
    pivot_data = grouped_data.pivot(index='year', columns='density class', values='Count').fillna(0)

    # Ensure columns are in the specified order
    pivot_data = pivot_data[density_order]

    # Define custom colors (lighter for low, darker for high)
    custom_colors = {
        'Very Low': '#56B4E9',  # Light Blue
        'Low': '#009E73',       # Green
        'Medium': '#F0E442',    # Yellow
        'High': '#E69F00',      # Orange
        'Very High': '#D55E00'  # Red
    }

    # Plot line graph with custom colors
    ax = pivot_data.plot(kind='line', figsize=(14, 8), color=[custom_colors[density] for density in density_order])

    # Add titles and labels
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Number of Occurrences')
    plt.legend(title='Density Class', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

    # Add gridlines for clarity
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
    plt.close('all')  # Clean up

# Example: Plot for each DataFrame
plot_density_class_over_time(neuston_net, title='Density Class Distribution for Neuston Net Method')
plot_density_class_over_time(hand_picking, title='Density Class Distribution for Hand Picking Method')
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd

def plot_density_class_over_time(data, title='Density Class Distribution Over Time'):
    """
    Plots a line graph of Density Class over time for a given dataset, with classes ordered from Very High to Very Low.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'year' and 'density class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the plot.
    """
    # Define the desired reversed order of density classes
    density_order = ['Very High', 'High', 'Medium', 'Low', 'Very Low']

    # Ensure 'density class' is a categorical column with the specified order
    data['density class'] = pd.Categorical(data['density class'], categories=density_order, ordered=True)

    # Group data by 'year' and 'Density Class', count occurrences
    grouped_data = data.groupby(['year', 'density class']).size().reset_index(name='Count')

    # Pivot the table for line plot
    pivot_data = grouped_data.pivot(index='year', columns='density class', values='Count').fillna(0)

    # Ensure columns are in the specified order
    pivot_data = pivot_data[density_order]

    # Define a gradient colormap (e.g., "viridis")
    colormap = cm.get_cmap('RdYlGn', len(density_order))  # Example: using "plasma" colormap
    gradient_colors = [colormap(i / (len(density_order) - 1)) for i in range(len(density_order))]

    # Plot line graph with gradient colors
    ax = pivot_data.plot(kind='line', figsize=(14, 8), color=gradient_colors)

    # Add titles and labels
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Number of Occurrences')
    plt.legend(title='Density Class', bbox_to_anchor=(1.05, 1), loc='upper left', labels=density_order)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

    # Add gridlines for clarity
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
    plt.close('all')  # Clean up

# Example: Plot for each DataFrame
plot_density_class_over_time(neuston_net, title='Density Class Distribution for Neuston Net Method')
plot_density_class_over_time(hand_picking, title='Density Class Distribution for Hand Picking Method')
plot_density_class_over_time(manta_net, title= 'Density Class Distribution for Manta Net Method Over Time')
plot_density_class_over_time(grab_sample, title= 'Density Class Distribution for Grab Sample Method Over Time')
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd

# Function for line plots: Density Class Over Time
def plot_density_class_over_time(data, title='Density Class Distribution Over Time'):
    """
    Plots a line graph of Density Class over time for a given dataset.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'year' and 'density class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the plot.
    """
    # Define the desired order of density classes
    density_order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

    # Ensure 'density class' is categorical and ordered
    data['density class'] = pd.Categorical(data['density class'], categories=density_order, ordered=True)

    # Group data by year and density class, count occurrences
    grouped_data = data.groupby(['year', 'density class']).size().reset_index(name='Count')

    # Pivot the data for line plots
    pivot_data = grouped_data.pivot(index='year', columns='density class', values='Count').fillna(0)

    # Ensure columns follow the desired order
    pivot_data = pivot_data[density_order]

    # Define colormap
    colormap = cm.get_cmap('viridis', len(density_order))  # Replace 'viridis' with other gradients if needed
    gradient_colors = [colormap(i / (len(density_order) - 1)) for i in range(len(density_order))]

    # Plot the data
    ax = pivot_data.plot(kind='line', figsize=(14, 8), color=gradient_colors)

    # Add titles and labels
    plt.title(title, fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Number of Occurrences', fontsize=14)
    plt.legend(title='Density Class', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    ax.grid(alpha=0.5)

    # Show the plot
    plt.show()
    plt.close('all')


# Function for longitude-latitude scatter plots
def plot_density_class_static_map(data, title='Density Class Distribution by Location'):
    """
    Plots a static scatter map of Density Class using Longitude and Latitude for a given dataset.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'longitude', 'latitude', and 'density class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the static map.
    """
    # Define the desired order of density classes
    density_order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

    # Ensure 'density class' is categorical and ordered
    data['density class'] = pd.Categorical(data['density class'], categories=density_order, ordered=True)

    # Define custom colors
    custom_colors = {
        'Very Low': '#56B4E9',  # Light Blue
        'Low': '#009E73',       # Green
        'Medium': '#F0E442',    # Yellow
        'High': '#E69F00',      # Orange
        'Very High': '#D55E00'  # Red
    }

    # Plot data
    plt.figure(figsize=(12, 8))
    for density in density_order:
        subset = data[data['density class'] == density]
        plt.scatter(
            subset['longitude'], 
            subset['latitude'], 
            label=density, 
            color=custom_colors[density], 
            alpha=0.8, 
            s=50  # Adjust size as needed
        )

    # Add titles and labels
    plt.title(title, fontsize=16)
    plt.xlabel('Longitude', fontsize=14)
    plt.ylabel('Latitude', fontsize=14)
    plt.legend(title='Density Class', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(alpha=0.5)

    # Show the plot
    plt.show()
    plt.close('all')


# Generate plots for all datasets
datasets = [
    (neuston_net, 'Density Class Distribution for Neuston Net Method'),
    (hand_picking, 'Density Class Distribution for Hand Picking Method'),
    (manta_net, 'Density Class Distribution for Manta Net Method'),
    (grab_sample, 'Density Class Distribution for Grab Sample Method')
]

# Over Time Plots
for data, title in datasets:
    plot_density_class_over_time(data, title=title + ' Over Time')

# Longitude-Latitude Maps
for data, title in datasets:
    plot_density_class_static_map(data, title=title + ' by Location')
```

```python
import plotly.express as px

def plot_density_class_animated_map(data, title='Density Class Distribution Over Time and Space'):
    """
    Plots an animated scatter map of Density Class using Longitude, Latitude, and Year for a given dataset.

    Parameters:
    - data (DataFrame): A pandas DataFrame containing 'longitude', 'latitude', 'year', and 'density class' columns.
    - title (str): Title of the plot.

    Returns:
    - None: Displays the interactive animated map.
    """
    # Define the desired order of density classes
    density_order = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

    # Ensure 'density class' is categorical with the specified order
    data['density class'] = pd.Categorical(data['density class'], categories=density_order, ordered=True)

    # Create the interactive animated map
    fig = px.scatter_geo(
        data,
        lon='longitude',
        lat='latitude',
        color='density class',
        animation_frame='year',  # Year-based animation
        category_orders={'density class': density_order},  # Ensure consistent density class order
        title=title,
        color_discrete_sequence=px.colors.sequential.Plasma,  # Use a sequential colormap
        hover_name='density class',  # Show density class on hover
        projection='natural earth',  # Map projection
    )

    # Customize layout
    fig.update_layout(
        legend_title_text='Density Class',
        title=dict(x=0.5),  # Center the title
        margin={"r": 0, "t": 50, "l": 0, "b": 0}  # Adjust margins
    )

    # Show the interactive animated map
    fig.show()


# Example: Animated Maps for Each Sampling Method
plot_density_class_animated_map(neuston_net, title='Density Class Distribution for Neuston Net Method Over Time')
plot_density_class_animated_map(hand_picking, title='Density Class Distribution for Hand Picking Method Over Time')
plot_density_class_animated_map(manta_net, title='Density Class Distribution for Manta Net Method Over Time')
plot_density_class_animated_map(grab_sample, title='Density Class Distribution for Grab Sample Method Over Time')
```
